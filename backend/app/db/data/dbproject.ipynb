```python
# Convert lastOrderID to string type first
customers['lastOrderID'] = customers['lastOrderID'].astype('string')

# Get the latest order ID for each customer
for customer in customers['id']:
    last_order = orders[orders['customerID'] == customer].sort_values(by='orderDate', ascending=False)
    if not last_order.empty:
        customers.loc[customers['id'] == customer, 'lastOrderID'] = last_order.iloc[0]['id']

customers
```

```python
# Create email addresses from customer names (lowercase, no spaces, @gmail.com)
customers['email'] = customers['name'].str.lower().str.replace(' ', '') + '@gmail.com'
customers
```

```python
# Define area codes for East Coast states
east_coast_area_codes = ['203', '207', '339', '603', '201', '212', '401', '215', '802']

def generate_phone():
    area = random.choice(east_coast_area_codes)  # Pick from East Coast area codes
    prefix = str(random.randint(100, 999))
    line = str(random.randint(1000, 9999))
    return f"({area}) {prefix}-{line}"

# Generate phone number for each customer
customers['phone'] = [generate_phone() for _ in range(len(customers))]
```
